## Task 1: Apply t-SNE

We will now apply t-Distributed Stochastic Neighbor Embedding (t-SNE) to reduce the dimensionality of our standardized Mall Customer data to 2 components. t-SNE is particularly good at visualizing high-dimensional data in a low-dimensional space, often revealing clusters that are not linearly separable.

```python
from sklearn.manifold import TSNE

# Step 1: Run t-SNE
# Ensure X_scaled is available from previous steps (loading and standardizing data)
# If not, please run the previous cells to define X_scaled.

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

print("Data transformed to 2 dimensions using t-SNE. First 5 rows:")
print(X_tsne[:5])
```

```python
# Step 2: Plot the result
plt.figure(figsize=(10, 8))
# For coloring by cluster labels, you would pass `c=cluster_labels` here
# Assuming 'cluster_labels' were derived from a previous clustering step (K-Means, DBSCAN) and are available.
# For now, we plot without specific cluster coloring as 'cluster_labels' are not defined in this run.
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], s=50, cmap='viridis', edgecolor='k')

plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('2D t-SNE Projection of Mall Customer Data')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Step 3: Compare to your PCA plot from Class 3 — are clusters more or less separated?

Comparing the t-SNE plot to the PCA plot:

*   **PCA (Principal Component Analysis)** is a linear dimensionality reduction technique that aims to maximize variance. It tries to project data onto a lower-dimensional subspace while preserving as much of the data's variance as possible. It is good for understanding the linear relationships and feature importance but might not effectively separate non-linearly separable clusters.

*   **t-SNE (t-Distributed Stochastic Neighbor Embedding)** is a non-linear dimensionality reduction technique that focuses on preserving local structures (i.e., preserving small pairwise distances or neighborhood identities). This often results in clusters appearing more distinct and separated in the low-dimensional space, even if they are not linearly separable in the original high-dimensional space.

In this case, the t-SNE plot likely shows more distinct, tighter clusters compared to the PCA plot, as t-SNE is designed to bring similar points together and push dissimilar points apart in the low-dimensional embedding, making the inherent grouping more visible. If you had cluster labels from a previous analysis, coloring the t-SNE plot would further highlight this separation.

### Comparison: PCA vs t-SNE vs UMAP

After visualizing the Mall Customer data with PCA, t-SNE, and UMAP, we can observe distinct characteristics of each method:

*   **PCA:** This is a linear method that focuses on preserving global variance. It often shows broad separations, but clusters might overlap significantly, especially if they are non-linearly separable. PCA is fast and its components can be interpreted (as we did in the previous section). It's excellent for initial data exploration and noise reduction, especially when linearity is assumed or desired.

*   **t-SNE:** This non-linear method excels at preserving local structures, meaning it tries to keep similar points close together and dissimilar points far apart. It often produces visually compelling clusters, appearing distinct and well-separated. However, it can be computationally intensive for very large datasets, and the distances between clusters in the t-SNE plot don't necessarily reflect actual distances in the high-dimensional space (don't over-interpret inter-cluster distances). It's best for visualizing inherent cluster structures.

*   **UMAP:** Similar to t-SNE, UMAP is also a non-linear technique designed for manifold learning. It generally offers a good balance between preserving local and global data structures. UMAP is often faster than t-SNE, scales better to larger datasets, and is better at preserving the global structure, which means that inter-cluster distances are more meaningful than in t-SNE. For many applications, UMAP provides visually appealing and more reliable representations than t-SNE.

**Which method gives the clearest picture of your customer segments?**

For the Mall Customer data, **t-SNE** and **UMAP** generally provide a clearer picture of potential customer segments compared to PCA. Both non-linear methods can reveal more intricate groupings that PCA might miss due to its linearity. UMAP often offers a slightly more balanced view by also trying to preserve global structure, making its output potentially more robust for understanding overall relationships between clusters.

**When to use each method:**

*   **PCA:** Use when speed is critical, interpretability of components is important, or when dealing with high-dimensional data that has mostly linear relationships. It's great for initial noise reduction and finding dominant directions of variance.
*   **t-SNE:** Use when the primary goal is high-quality visualization of inherent clusters, especially for non-linear structures. Be mindful of its computational cost and the interpretability of inter-cluster distances.
*   **UMAP:** A strong alternative to t-SNE, offering competitive visualization quality (often clearer and more consistent) but with better speed and global structure preservation. It's a good general-purpose tool for high-dimensional data visualization and exploration when both local and global structures are important.
